```python
import pandas as pd

# 1. ÌååÏùº Î∂àÎü¨Ïò§Í∏∞
house_df = pd.read_csv("C:/Users/hgj04/OneDrive/Î¨∏ÏÑú/Ïπ¥Ïπ¥Ïò§ÌÜ° Î∞õÏùÄ ÌååÏùº/Í∞ÄÍµ¨Ï£ºÌÉù.csv", encoding="cp949")
people_df = pd.read_csv("C:/Users/hgj04/OneDrive/Î¨∏ÏÑú/Ïπ¥Ïπ¥Ïò§ÌÜ° Î∞õÏùÄ ÌååÏùº/Ïù∏Íµ¨Ïó∞Î†π1.csv", encoding="cp949")

# 2. ÏûêÎÖÄ Ïàò Í≥ÑÏÇ∞
people_df["ÏûêÎÖÄÏàò"] = people_df["ÎÇ®ÏïÑÏ∂úÏÇ∞Ïàò"] + people_df["Ïó¨ÏïÑÏ∂úÏÇ∞Ïàò"]
children_per_house = people_df.groupby("Í∞ÄÍµ¨ÏùºÎ†®Î≤àÌò∏")["ÏûêÎÖÄÏàò"].sum().reset_index()

# ‚úÖ 3. Í∞ÄÍµ¨Ï£º(1Î≤à) + Í∞ÄÍµ¨ÏõêÎ≤àÌò∏ 2Î≤àÏùò Ïó∞Î†π ÌèâÍ∑† Í≥ÑÏÇ∞
relevant_people = people_df[people_df["Í∞ÄÍµ¨ÏõêÎ≤àÌò∏"].isin([1, 2])]
householder_avg_age = relevant_people.groupby("Í∞ÄÍµ¨ÏùºÎ†®Î≤àÌò∏")["ÎßåÏó∞Î†π"].mean().reset_index()

# 4. Í∞ÄÍµ¨ÏûêÎ£åÏôÄ Î≥ëÌï©
merged_df = pd.merge(house_df, children_per_house, on="Í∞ÄÍµ¨ÏùºÎ†®Î≤àÌò∏", how="left")
merged_df["ÏûêÎÖÄÏàò"] = merged_df["ÏûêÎÖÄÏàò"].fillna(0)

merged_df = pd.merge(merged_df, householder_avg_age, on="Í∞ÄÍµ¨ÏùºÎ†®Î≤àÌò∏", how="left")

# 5. ÏΩîÎìú Îß§Ìïë
ownership_map = {1: "ÏûêÍ∞Ä", 2: "Ï†ÑÏÑ∏", 3: "ÏõîÏÑ∏", 4: "Î¨¥ÏÉÅ", 5: "Í∏∞ÌÉÄ"}
housing_type_map = {
    1: "Îã®ÎèÖÏ£ºÌÉù", 2: "ÏïÑÌååÌä∏", 3: "Ïó∞Î¶ΩÏ£ºÌÉù", 4: "Îã§ÏÑ∏ÎåÄÏ£ºÌÉù",
    5: "ÎπÑÍ±∞Ï£ºÏö© Í±¥Î¨º ÎÇ¥ Ï£ºÌÉù", 6: "Ï£ºÌÉù Ïù¥Ïô∏Ïùò Í±∞Ï≤ò", 7: "Í∏∞ÌÉÄ"
}

merged_df["Ï£ºÍ±∞ÌòïÌÉú"] = merged_df["Ï†êÏú†ÌòïÌÉúÏΩîÎìú"].map(ownership_map)
merged_df["Ï£ºÌÉùÏ¢ÖÎ•ò"] = merged_df["Í±∞Ï≤òÏ¢ÖÎ•òÏΩîÎìú"].map(housing_type_map)

# 6. Ïó∞Î†πÎåÄ Î≥ÄÏàò Ï∂îÍ∞Ä (ÌèâÍ∑† ÎÇòÏù¥Ïóê Îî∞Îùº)
merged_df["Ïó∞Î†πÎåÄ"] = pd.cut(
    merged_df["ÎßåÏó∞Î†π"],
    bins=[19, 29, 39, 49, 59],
    labels=["20ÎåÄ", "30ÎåÄ", "40ÎåÄ", "50ÎåÄ"],
    include_lowest=True
)

filtered_df = merged_df[merged_df["Ïó∞Î†πÎåÄ"].notna()]

# 7. Î∂ÑÏÑù
result = merged_df.groupby(["Ïó∞Î†πÎåÄ", "Ï£ºÍ±∞ÌòïÌÉú"], observed=True)["ÏûêÎÖÄÏàò"].mean().round(2).unstack()
print("üìä Ïó∞Î†πÎåÄÎ≥Ñ Ï£ºÍ±∞ÌòïÌÉúÏóê Îî∞Î•∏ ÌèâÍ∑† ÏûêÎÖÄ Ïàò:")
print(result)

result_housing = merged_df.groupby(["Ïó∞Î†πÎåÄ", "Ï£ºÌÉùÏ¢ÖÎ•ò"], observed=True)["ÏûêÎÖÄÏàò"].mean().round(2).unstack()
print("\nüèòÔ∏è Ïó∞Î†πÎåÄÎ≥Ñ Ï£ºÌÉùÏ¢ÖÎ•òÏóê Îî∞Î•∏ ÌèâÍ∑† ÏûêÎÖÄ Ïàò:")
print(result_housing)
```
