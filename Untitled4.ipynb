```python
import pandas as pd

# 1. 파일 불러오기
house_df = pd.read_csv("C:/Users/hgj04/OneDrive/문서/카카오톡 받은 파일/가구주택.csv", encoding="cp949")
people_df = pd.read_csv("C:/Users/hgj04/OneDrive/문서/카카오톡 받은 파일/인구연령1.csv", encoding="cp949")

# 2. 자녀 수 계산
people_df["자녀수"] = people_df["남아출산수"] + people_df["여아출산수"]
children_per_house = people_df.groupby("가구일련번호")["자녀수"].sum().reset_index()

# ✅ 3. 가구주(1번) + 가구원번호 2번의 연령 평균 계산
relevant_people = people_df[people_df["가구원번호"].isin([1, 2])]
householder_avg_age = relevant_people.groupby("가구일련번호")["만연령"].mean().reset_index()

# 4. 가구자료와 병합
merged_df = pd.merge(house_df, children_per_house, on="가구일련번호", how="left")
merged_df["자녀수"] = merged_df["자녀수"].fillna(0)

merged_df = pd.merge(merged_df, householder_avg_age, on="가구일련번호", how="left")

# 5. 코드 매핑
ownership_map = {1: "자가", 2: "전세", 3: "월세", 4: "무상", 5: "기타"}
housing_type_map = {
    1: "단독주택", 2: "아파트", 3: "연립주택", 4: "다세대주택",
    5: "비거주용 건물 내 주택", 6: "주택 이외의 거처", 7: "기타"
}

merged_df["주거형태"] = merged_df["점유형태코드"].map(ownership_map)
merged_df["주택종류"] = merged_df["거처종류코드"].map(housing_type_map)

# 6. 연령대 변수 추가 (평균 나이에 따라)
merged_df["연령대"] = pd.cut(
    merged_df["만연령"],
    bins=[19, 29, 39, 49, 59],
    labels=["20대", "30대", "40대", "50대"],
    include_lowest=True
)

filtered_df = merged_df[merged_df["연령대"].notna()]

# 7. 분석
result = merged_df.groupby(["연령대", "주거형태"], observed=True)["자녀수"].mean().round(2).unstack()
print("📊 연령대별 주거형태에 따른 평균 자녀 수:")
print(result)

result_housing = merged_df.groupby(["연령대", "주택종류"], observed=True)["자녀수"].mean().round(2).unstack()
print("\n🏘️ 연령대별 주택종류에 따른 평균 자녀 수:")
print(result_housing)
```
